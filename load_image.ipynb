```python
from skimage import transform

original_image = io.imread("C:/Users/HOME/Pictures/Saved Pictures/3.jpg", as_gray = False)

resized_image = transform.resize(original_image, (300,300))

plt.subplot(121)
plt.imshow(original_image)
plt.title("Original image")

plt.subplot(122)
plt.imshow(resized_image)
plt.title("Resized image")

plt.show()
```

```python
from skimage import transform

original_image = io.imread("C:/Users/HOME/Pictures/Saved Pictures/5.jpg", as_gray = False)

rescaled_image = transform.rescale(original_image, (2.5, 0.5))

plt.subplot(121)
plt.imshow(original_image)
plt.title("Original image")

plt.subplot(122)
plt.imshow(rescaled_image)
plt.title("Rescaled image")

plt.show()
```

```python
from skimage import transform

original_image = io.imread("C:/Users/HOME/Pictures/Saved Pictures/7.jpg", as_gray = False)

rotated_image = transform.rotate(original_image, angle = 45, resize = True)

plt.subplot(121)
plt.imshow(original_image)
plt.title("Original image")

plt.subplot(122)
plt.imshow(rotated_image)
plt.title("Rotated image")

plt.show()
```
